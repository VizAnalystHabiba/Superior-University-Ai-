```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Set up the figure with 5 rows and 2 columns
fig, axes = plt.subplots(5, 2, figsize=(15, 20))
fig.tight_layout(pad=5.0)

# Plotting different graphs in the subplots
# Row 1, Col 1
axes[0, 0].plot(x, y, label='Sine Wave')
axes[0, 0].set_title('Line Plot')
axes[0, 0].legend()

# Row 1, Col 2
sns.histplot(y, bins=20, ax=axes[0, 1])
axes[0, 1].set_title('Histogram')

# Row 2, Col 1
axes[1, 0].scatter(x, y, color='r')
axes[1, 0].set_title('Scatter Plot')

# Row 2, Col 2
sns.boxplot(x=y, ax=axes[1, 1])
axes[1, 1].set_title('Box Plot')

# Row 3, Col 1
sns.lineplot(x=x, y=np.cos(x), ax=axes[2, 0])
axes[2, 0].set_title('Cosine Wave')

# Row 3, Col 2
axes[2, 1].bar([1, 2, 3, 4], [10, 15, 7, 10])
axes[2, 1].set_title('Bar Chart')

# Row 4, Col 1
sns.violinplot(x=y, ax=axes[3, 0])
axes[3, 0].set_title('Violin Plot')

# Row 4, Col 2
axes[3, 1].fill_between(x, y, color="skyblue", alpha=0.4)
axes[3, 1].set_title('Area Plot')

# Row 5, Col 1
axes[4, 0].plot(x, np.tan(x), label='Tangent Wave', color='g')
axes[4, 0].set_title('Tangent Plot')
axes[4, 0].legend()

# Row 5, Col 2
sns.kdeplot(x=y, ax=axes[4, 1], shade=True)
axes[4, 1].set_title('KDE Plot')

# Display the plot
plt.show()
```

```python
ax = sns.countplot(data=df, x='Survived')
plt.title('Survival Distribution (0=Dead, 1=Survived)')

# For adding labels
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'center', 
                xytext = (0, 9), 
                textcoords = 'offset points')

plt.show()
```

```python
# Family Size Analysis
df['Family_Size'] = df['SibSp'] + df['Parch']
plt.figure(figsize=(10, 6))
sns.countplot(x='Family_Size', hue='Survived', data=df)
plt.title('Survival by Family Size')
plt.show()
```

```python
#  Fare Distribution by Class
plt.figure(figsize=(10, 6))
sns.boxplot(x='Pclass', y='Fare', data=df)
plt.title('Fare Distribution by Class')
plt.show()
```
