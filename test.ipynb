```python
# Download historical data for Apple (AAPL)
ticker = 'AAPL'
data = yf.download(ticker, start='2020-01-01', end='2024-10-07')

# Display the first few rows of the dataset
print(data.head())
```

```python
# Plotting the closing price of AAPL
plt.figure(figsize=(14, 7))
plt.plot(data['Close'], label='AAPL Closing Price', color='blue')
plt.title('Apple Stock Closing Prices (2020-2024)')
plt.xlabel('Date')
plt.ylabel('Closing Price (USD)')
plt.legend()
plt.show()
```

```python
# Calculate the moving average
data['MA50'] = data['Close'].rolling(window=50).mean()
plt.figure(figsize=(14, 7))
plt.plot(data['Close'], label='AAPL Closing Price', color='blue')
plt.plot(data['MA50'], label='50-Day Moving Average', color='orange')
plt.title('Apple Stock Closing Prices with 50-Day Moving Average')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.show()
```

```python
import matplotlib.dates as mdates

# Plotting the trading volume with modifications
plt.figure(figsize=(14, 5))
colors = ['red' if x > data['Volume'].median() else 'green' for x in data['Volume']]
plt.bar(data.index, data['Volume'], color=colors, alpha=0.7)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.title('Trading Volume for Apple (AAPL)', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Volume', fontsize=14)
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.xticks(rotation=45)
plt.tight_layout() 
plt.show()
```
